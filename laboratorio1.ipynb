### Análisis de Componentes Principales

Para determinar si es conveniente hacer un análisis de componentes principales, se deben evaluar los indicadores con el test KMO (Kaiser–Meyer–Olkin) y el test de esfericidad de Barlett.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
```

```python
variables_cuantitativas = [
    'Age', 'Number of sexual partners', 'First sexual intercourse',
    'Num of pregnancies', 'Smokes (years)', 'Smokes (packs/year)',
    'Hormonal Contraceptives (years)', 'IUD (years)', 'STDs (number)',
    'STDs: Number of diagnosis'
]
df = df[variables_cuantitativas].copy()
df
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled
```

#### Calcular KMO TEST

```python
kmo_all, kmo_model = calculate_kmo(df_scaled)
print(f"KMO: {kmo_model}")
```

```python
bartlett_test = calculate_bartlett_sphericity(df_scaled)
print(f"Chi-square: {bartlett_test[0]}")
print(f"p-value: {bartlett_test[1]}")
```

El índice KMO, con un valor menor a 0.5 sugiere que la realización del PCA sea adecuada. En el caso de este dataset, se obtuvo un valor de 0.5244, por lo que si indica que se podría hacer el PCA.
Cuando el p-value del test de Bartlett es 0.0, indica que las variables no están correlacionadas, por lo que hay conflicto entre hacer el PCA debido a los resultados distintos entre el KMO y el test de Bartlett.

Por fines de realizar este laboratorio, se procederá a realizar el PCA.

```python
pca = PCA(n_components=len(variables_cuantitativas))
pca.fit(df_scaled)
```

```python
# Get the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Get the cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Print the explained variance ratio and cumulative explained variance ratio
for i in range(len(explained_variance_ratio)):
    print(f"Principal Component {i+1}: Explained Variance Ratio = {explained_variance_ratio[i]:.4f}, Cumulative Explained Variance Ratio = {cumulative_variance_ratio[i]:.4f}")
```

```python
import matplotlib.pyplot as plt

# Obtener la varianza explicada por cada componente
explained_variance = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_

# Crear el gráfico de barras
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5)
plt.plot(range(1, len(explained_variance_ratio) + 1), np.cumsum(explained_variance_ratio), marker='o', color='r')
plt.xlabel('Componente Principal')
plt.ylabel('Varianza Explicada')
plt.title('Varianza Explicada por Componente Principal')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.show()
```

Es importante tener en cuenta que la interpretación de los componentes puede variar dependiendo del contexto y la naturaleza de los datos. Se recomienda realizar un análisis más detallado y considerar el conocimiento del dominio para una interpretación más precisa.

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming 'df' is your DataFrame and 'variables_cuantitativas' contains the names of numerical columns
variables_cuantitativas = ['Age', 'Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies', 'Smokes (years)', 'Smokes (packs/year)', 'Hormonal Contraceptives (years)', 'IUD (years)', 'STDs (number)', 'STDs: Number of diagnosis']

# Discretize numerical variables
df_discretized = df.copy()

# Ensure the columns are numeric and handle missing values
for var in variables_cuantitativas:
    df_discretized[var] = pd.to_numeric(df_discretized[var], errors='coerce')  # Convert to numeric, coerce errors to NaN
    df_discretized[var].fillna(-1, inplace=True)  # Fill NaNs with a placeholder value (e.g., -1)

# Bin the numeric columns
for var in variables_cuantitativas:
    df_discretized[f'{var}_binned'] = pd.cut(df_discretized[var], bins=5, labels=False)

# Convert dataset into a transaction format
transactions = df_discretized.apply(lambda row: [str(row[col]) for col in df_discretized.columns], axis=1).tolist()

# Perform association rule mining
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df_transactions, min_support=0.1, use_colnames=True)
association_rules_df = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Print the association rules
print(association_rules_df)
```
