##### Histogramas

```python
import matplotlib.pyplot as plt

# Histograma para variables numéricas
df.hist(bins=30, figsize=(20, 15))
plt.show()
```

Como se observa en los histogramas, hay variables que tienen únicamente valores de 0 o 1. Esto indica que estas variables son entre sí o no, lo cual dice que son las variables categóricas. Además, por el contexto del nombre de la variable (como "Smokes" = "Fuma) se puede asegurar que este es el caso.

##### Gráficos para variables categóricas

```python
# Gráficos de barras para variables categóricas

categorical_cols = [col for col in df.columns if set(df[col].unique()) == {0, 1}]

for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=col, data=df)
    plt.xticks(rotation=90)
    plt.show()
```

Como se observa en los gráficos de las variables categóricas, la mayoría de personas no tienen ninguna de las STDs y tampoco se han hecho exámenes. Además, se puede observar que la gente SÍ toma contraceptivos hormonales.

```python
# Aislar las variables numéricas
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Matriz de correlación
correlation_matrix = df[numerical_cols].corr()

# Mapa de calor de la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

Según el gráfico de correlación que se ha generado, la mayoría de correlación entre variables numéricas existe entre aquellas enfermedades que son iguales. Como se puede observar con las STDs (ETS) tienen correlación entre los diagnósticos, el número, entre otras variables.

# Transformaciones de Variables Categóricas para PCA: ¿Vale la Pena?

## 1. Entendiendo el PCA
- Reduce dimensionalidad de datos numéricos, transformándolos en componentes no correlacionados.

## 2. Transformaciones de Variables Categóricas
- **Codificación One-Hot:** Categorías a columnas binarias.
- **Codificación de Etiquetas:** Números a categorías (puede introducir falsas relaciones).

## 3. Ventajas y Desventajas

### Ventajas
- **Captura de Variabilidad:** Mejora la interpretación si las categorías son relevantes.

### Desventajas
- **Aumento de Dimensionalidad:** La codificación one-hot puede ser contraproducente.
- **Relaciones Espurias:** La codificación de etiquetas puede ser engañosa.

## 4. ¿Vale la Pena?
- Depende del conjunto de datos y la relevancia de las categorías.
- Realizar pruebas con y sin categorías para evaluar el impacto.
